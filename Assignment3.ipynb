```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm 
#%matplotlib 
```

```python
#this notebook takes a while to run, please be patient, it will execute eventually, thank you!

#generate list of elements 1 to 2024
list_2024 = np.arange(1,2025)


#creating a function to play a game with a given list
def game(List):
    
    while len(List) > 1:
   
        #picking a random index for the length of the list
        a = random.randint(0, len(List)-1)

        #defining what number is at that index 
        num1 = List[a]
        
        #deleting the picked number from the list
        List = np.delete(List, a)
        
        #picking a random index for the length of the list
        b = random.randint(0, len(List)-1)

        #defining what the number is at that index 
        num2 = List[b]
        
        #deleting the second picked number from the list
        List = np.delete(List, b)
        
        #calculating the positive difference of the two picked numbers
        diff = abs(num1 - num2)
        
        #appending the difference to the list
        List = np.append(List, diff)
     
    return List[0]    
  
print(game(list_2024))
```

```python
#defining the list of numbers 1 to 2023
list_2023 = np.arange(1, 2024)

#creating a while loop to keep playing games until 0 is the final number
while game(list_2023) != 0:
    game(list_2023)
    
print('the game with 2023 as the starting number, has 0 as a final number')
```

```python
#creating a list of final numbers when playing the game multiple times
Final = []

plt.figure(figsize = (9,3))

for i in range(3000):
    c = game(list_2024)
    Final.append(c)
    plt.scatter(c, 0, marker = 'x')

plt.xlabel('Final Number')
plt.title('Plotting the Final Numbers')
plt.show()
```

```python
#plotting the histogram of the final numbers data
plt.figure(figsize = (9,6))
sns.histplot(Final, bins=100, stat = 'density')

#plotting the normal distribution of the final numbers data
x2 = np.linspace(min(Final), max(Final))
MU = np.mean(Final)
SIG = np.sqrt(np.var(Final, ddof=1))
y2 = norm.pdf(x2, MU, SIG)
plt.plot(x2,y2, 'r')
plt.xlabel('Final Number')
plt.title('Histogram of Final Numbers When Playing 3000 Games, Compared to the Normal Distribution of the Data')
plt.grid()
plt.show()
```
